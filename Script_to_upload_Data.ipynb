```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/sumee/OneDrive/Desktop/dataset.csv")
```

```python
df.head()
```

```python
df.drop('Instructions',axis=1,inplace=True)
```

```python
df.head()
```

```python
for index in df.index:
    if 'Vegetarian' in df['Diet'][index] and 'Non' not in df['Diet'][index]:
        df['Diet'][index] = 'V'
    elif 'Vegetarian' in df['Diet'][index] and 'Non' in df['Diet'][index]:
        df['Diet'][index] = 'NV'
    elif 'Vegeterian' in df['Diet'][index] and 'Non' in df['Diet'][index]:
        df['Diet'][index] = 'NV'
    elif 'Eggetarian' in df['Diet'][index]:
        df['Diet'][index] = 'EV'
```

```python
df['Diet'].replace('Vegan','V',inplace=True)
```

```python
df['Diet'].unique()
```

```python
df.to_csv("Dataset")
```

```python
import requests
url = "https://software-engineering-1.onrender.com/apis/v1/Recipe/"
```

```python
for index in df.index:
    row = {'name':df['Translated Recipe Name'][index],'ingredients':df['TranslatedIngredients'][index],'type':df['Diet'][index],'instructions':df['TranslatedInstructions'][index],'course':df['Course'][index],'website':df['URL'][index]}
    response = requests.post(url=url,data=row)
    print(response.status_code)
```
